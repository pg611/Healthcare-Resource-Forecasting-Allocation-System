# Modelling with Stratified K Fold cross validation

1. StratifiedKFold splits the data into n_splits (here, 5) folds, ensuring that the class distribution in each fold is similar to the original dataset.
2. This prevents issues like over-representation or under-representation of certain classes in the training or validation sets.

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score
from lightgbm import LGBMClassifier
import numpy as np

accuracy = []
final_preds = []
folds = StratifiedKFold(n_splits=5)

for train_index, test_index in folds.split(X, y):
    # Creating training and validation datasets
    X_Train, X_Test = X.iloc[train_index], X.iloc[test_index]
    y_Train, y_Test = y.iloc[train_index], y.iloc[test_index]
    
    # Building a classifier
    clf = LGBMClassifier(
        n_estimators=750,
        learning_rate=0.1,
        objective="multiclass",
        boosting_type="gbdt",
        subsample=0.9,
        colsample_bytree=0.6,
        num_class=11,
        max_depth=12,
        n_jobs=-1,
        reg_alpha=2,
        num_leaves=100
    )
    
    # Fitting the classifier to the train set
    clf.fit(
        X_Train,
        y_Train,
        eval_set=[(X_Train, y_Train), (X_Test, y_Test)],
        eval_metric="multi_logloss"  
    )
    
    # Predicting on the validation set
    preds = clf.predict(X_Test)
    score = accuracy_score(preds, y_Test)
    print("Accuracy Score:", score)
    print("------------------------------------------------")
    accuracy.append(score)
    
    # Predicting on the test set
    pred = clf.predict(test)
    final_preds.append(pred)

print("------------------------------------------")
print("Mean Accuracy of 5 Folds:", np.mean(np.array(accuracy)))
```

# Resource Allocation

Aggregate Predicted Stays by Hospital

```python
# Ensure "Department" exists in the test DataFrame (if not, add it from train or other sources)
if "Department" not in test.columns:
    test["Department"] = train["Department"][:len(test)]  # Example to add from train if test lacks it

# Map predicted stays to approximate patient-days
test["Predicted_Stay"] = Submission['Stay'].map({
    "0-10": 5, "11-20": 15, "21-30": 25, "31-40": 35, 
    "41-50": 45, "51-60": 55, "61-70": 65, "71-80": 75, 
    "81-90": 85, "91-100": 95, "More than 100 Days": 110
})

# Aggregate resource demand at hospital and department levels
hospital_resource_demand = test.groupby(["Hospital_code", "Department"]).agg({
    "Predicted_Stay": "sum",  # Total predicted patient-days
    "Extra_Rooms": "first"    # Available extra rooms (assuming constant per hospital)
}).reset_index()

# Calculate required resources
hospital_resource_demand["Required_Rooms"] = (
    hospital_resource_demand["Predicted_Stay"] / 10
).round()
hospital_resource_demand["Allocated_Rooms"] = hospital_resource_demand[
    ["Required_Rooms", "Extra_Rooms"]
].min(axis=1)

# Add additional resource needs
hospital_resource_demand['Required_Nurses'] = hospital_resource_demand['Predicted_Stay'] // 5
hospital_resource_demand['Required_Doctors'] = hospital_resource_demand['Predicted_Stay'] // 20

# Save results to CSV
hospital_resource_demand.to_csv("resource_allocation.csv", index=False)

# Display sample output
print(hospital_resource_demand.head())
```

1. Surgery Departments typically handle cases requiring longer patient stays (e.g., recovery after surgery, post-operative care).

```python
import numpy as np
import matplotlib.pyplot as plt

# Aggregate the data by Hospital Code
hospital_aggregated = hospital_resource_demand.groupby("Hospital_code").agg({
    "Required_Rooms": "sum",
    "Allocated_Rooms": "sum"
}).reset_index()

# Normalize the aggregated values
hospital_aggregated["Required_Rooms_Norm"] = hospital_aggregated["Required_Rooms"] / hospital_aggregated["Required_Rooms"].max()
hospital_aggregated["Allocated_Rooms_Norm"] = hospital_aggregated["Allocated_Rooms"] / hospital_aggregated["Allocated_Rooms"].max()

# Plot normalized values
plt.figure(figsize=(14, 8))
bar_width = 0.4
x_positions = np.arange(len(hospital_aggregated))

plt.bar(
    x_positions, 
    hospital_aggregated["Required_Rooms_Norm"], 
    width=bar_width, 
    color="skyblue", 
    label="Required Rooms (Normalized)"
)
plt.bar(
    x_positions + bar_width, 
    hospital_aggregated["Allocated_Rooms_Norm"], 
    width=bar_width, 
    color="green", 
    label="Allocated Rooms (Normalized)"
)

plt.title("Comparison of Normalized Required vs Allocated Rooms by Hospital Code", fontsize=16)
plt.xlabel("Hospital Code", fontsize=12)
plt.ylabel("Normalized Room Count", fontsize=12)
plt.xticks(x_positions + bar_width / 2, hospital_aggregated["Hospital_code"], rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```
